```python
import numpy as np
```

```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("xray.jpeg")
plt.imshow(img)
```

```python
img =cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
```

```python
img =cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
sobelX = cv2.Sobel(img,cv2.CV_64F,1,0)
sobelY = cv2.Sobel(img,cv2.CV_64F,0,1)
```

```python
sobelX = np.uint8(np.absolute(sobelX))
sobelY = np.uint8(np.absolute(sobelY))
```

```python
combind = cv2.bitwise_or(sobelX,sobelY)
```

```python
lab =cv2.Laplacian(img,cv2.CV_64F,ksize = 3)
lab = np.uint8(np.absolute(lab))
```

```python
titles = ['image','sobelX','sobelY','Combined','Laplacian']
images = [img,sobelX,sobelY,combind,lab]

for i in range(5):
    plt.subplot(2,3,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
```
